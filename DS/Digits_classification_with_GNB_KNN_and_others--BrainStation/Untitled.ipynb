```python
def naive_cords(image):
    #df = pd.DataFrame(list(image.getdata()), columns=[filename])
    start_list = list(image.getdata())
    err1, err2 = 3, 4
    first_black, last_black = (err2*30+err2), (900-(err2*30+err2))

    counter = first_black
    for c in start_list[first_black:]:
        if (c==0):
            first_black = counter
            break
        counter+=1
    counter = last_black
    for c in start_list[:last_black:-1]:
        if (c==0):
            last_black = counter
            break
        counter-=1
    del counter
    
    fst_l_s = (10-err2)*(30) + err1
    fst_l_b = (20+err2)*(30) + err1
    lst_r_s = (10-err2)*(30) - err1 #- 1
    lst_r_b = (20+err2)*(30) - err1 #- 1
    fst_t_s = (30*err1) + (10 - err2)
    fst_t_b = (30*err1) + (20 + err2)
    lst_b_s = 30*(30-err1) + (10 - err2)
    lst_b_b = 30*(30-err1) + (20 + err2)
    
    left, right, top, bottom = 0, 0, 0, 0
    
    #counter1, counter2 = fst_l_s, (fst_l_s+30*err2//2
    counter = err1 
    for pixel in range(err1, 30-err1):
        a = True if (start_list[fst_l_s + 30*(-4) + counter - err1] == 0) else False
        b = True if (start_list[fst_l_s + 30*(-2) + counter - err1] == 0) else False
        c = True if (start_list[fst_l_b + 30*(-4) + counter - err1] == 0) else False
        d = True if (start_list[fst_l_b + 30*(-2) + counter - err1] == 0) else False
        e = True if (start_list[fst_l_s + 30*2 + counter - err1] == 0) else False
        f = True if (start_list[fst_l_s + 30*4 + counter - err1] == 0) else False
        g = True if (start_list[fst_l_b + 30*2 + counter - err1] == 0) else False
        h = True if (start_list[fst_l_b + 30*4 + counter - err1] == 0) else False
        if (counter > (30-err1)):
            break
        if ((a or b) or (c or d)):
            left = counter
            break
        if ((e or f) or (g or h)):
            left = counter
            break
        counter+=1
    
    counter = 30-err1 
    for pixel in range(err1, 30-err1):
        a = True if (start_list[lst_r_s + 30*(-4) + counter - 30 - err1] == 0) else False
        b = True if (start_list[lst_r_s + 30*(-2) + counter - 30 - err1] == 0) else False
        c = True if (start_list[lst_r_b + 30*(-4) + counter - 30 - err1] == 0) else False
        d = True if (start_list[lst_r_b + 30*(-2) + counter - 30 - err1] == 0) else False
        e = True if (start_list[lst_r_s + 30*2 + counter - 30 - err1] == 0) else False
        f = True if (start_list[lst_r_s + 30*4 + counter - 30 - err1] == 0) else False
        g = True if (start_list[lst_r_b + 30*2 + counter - 30 - err1] == 0) else False
        h = True if (start_list[lst_r_b + 30*4 + counter - 30 - err1] == 0) else False
        if (counter < err1):
            break
        if ((a or b) or (c or d)):
            right = counter
            break
        if ((e or f) or (g or h)):
            right = counter
            break
        counter-=1
        
    counter = err1 
    for pixel in range(err1, 30-err1):
        a = True if (start_list[fst_t_s - 4 + 30*(counter - err1)] == 0) else False
        b = True if (start_list[fst_t_s - 2 + 30*(counter - err1)] == 0) else False
        c = True if (start_list[fst_t_b - 4 + 30*(counter - err1)] == 0) else False
        d = True if (start_list[fst_t_b - 2 + 30*(counter - err1)] == 0) else False
        e = True if (start_list[fst_t_s + 2 + 30*(counter - err1)] == 0) else False
        f = True if (start_list[fst_t_s + 4 + 30*(counter - err1)] == 0) else False
        g = True if (start_list[fst_t_b + 2 + 30*(counter - err1)] == 0) else False
        h = True if (start_list[fst_t_b + 4 + 30*(counter - err1)] == 0) else False
        if (counter > (30-err1)):
                break
        if ((a or b) or (c or d)):
            top = counter
            break
        if ((e or f) or (g or h)):
            top = counter
            break
        counter+=1
        
    counter = 30-err1 
    for pixel in range(err1, 30-err1):
        a = True if (start_list[lst_b_s - 4 + 30*(counter - 30 - err1)] == 0) else False
        b = True if (start_list[lst_b_s - 2 + 30*(counter - 30 - err1)] == 0) else False
        c = True if (start_list[lst_b_b - 4 + 30*(counter - 30 - err1)] == 0) else False
        d = True if (start_list[lst_b_b - 2 + 30*(counter - 30 - err1)] == 0) else False
        e = True if (start_list[lst_b_s + 2 + 30*(counter - 30 - err1)] == 0) else False
        f = True if (start_list[lst_b_s + 4 + 30*(counter - 30 - err1)] == 0) else False
        g = True if (start_list[lst_b_b + 2 + 30*(counter - 30 - err1)] == 0) else False
        h = True if (start_list[lst_b_b + 4 + 30*(counter - 30 - err1)] == 0) else False
        if (counter < err1):
                break
        if ((a or b) or (c or d)):
            bottom = counter
            break
        if ((e or f) or (g or h)):
            bottom = counter
            break
        counter-=1
        
    return left, top, right, bottom
```

```python
def naive_cords2(image):
    #df = pd.DataFrame(list(image.getdata()), columns=[filename])
    start_list = list(image.getdata())
    err1 = 3
        
    left_pixel = []
    top_pixel = []
    right_pixel = []
    bottom_pixel = []
    arr_rng = range(err1, 30-err1)
    
    for count in arr_rng:
        left_pixel.append([0,0])
        top_pixel.append([0,0])
        right_pixel.append([0,0])
        bottom_pixel.append([0,0])
            
    #counter1, counter2 = fst_l_s, (fst_l_s+30*err2//2
    #counter_main = err1 
    for search in arr_rng:
        counter = err1
        for item in left_pixel:
            item[0] = search if ((start_list[counter*30 + search] == 0) and (item[0] == 0)) else item[0]
            item[1] += 1 if (start_list[counter*30 + search] == 0) else item [1]
            counter+=1
        
        counter = err1
        for item in top_pixel:
            item[0] = search if ((start_list[search*30 + counter] == 0) and (item[0] == 0)) else item[0]
            item[1] += 1 if (start_list[search*30 + counter] == 0) else item [1]
            counter+=1
        
        counter = err1
        for item in right_pixel:
            item[0] = (30 - search) if ((start_list[counter*30 + 30 - search] == 0) and (item[0] == 0)) else item [0]
            item[1] += 1 if (start_list[counter*30 + 30 - search] == 0) else item [1]
            counter+=1
            
        counter = err1
        for item in bottom_pixel:
            item[0] = (30 - search) if ((start_list[(30-search)*30 + counter] == 0) and (item[0] == 0)) else item [0]
            item[1] += 1 if (start_list[counter*30 + search] == 0) else item [1]
            counter+=1
            
    left, top, right, bottom = 0, 0, 0, 0
    
    left = 30 - err1;
    for i in left_pixel:
        left = i[0] if ((i[0] < left) and i[0] > 0) else left
    
    top = 30 - err1;
    for i in top_pixel:
        top = i[0] if ((i[0] < top) and i[0] > 0) else top
        
    right = err1;
    for i in right_pixel:
        right = i[0] if ((i[0] > right) and i[0] > 0) else right
    
    bottom = err1;
    for i in bottom_pixel:
        bottom = i[0] if ((i[0] > bottom) and i[0] > 0) else bottom
          
    return left, top, right, bottom
```

```python
def resizeim(image):
    temp_image = image.convert('1')
    #factor = (32 / temp_image.height) if (temp_image.height > temp_image.width) else (32 / temp_image.width)
    factorh = 30 / temp_image.height
    factorw = 30 / temp_image.width
    temp_image = temp_image.resize((round(temp_image.height * factorh), round(temp_image.width * factorw)))
    fin_image = Image.new('1', (30,30), 'white')
    #x = round(15 - (int((x2+1)*factor) - (int((x1-1)*factor)))/2)
    #y = round(15 - (int((y2+1)*factor) - (int((y1-1)*factor)))/2)
    x = round((30 - temp_image.width)/2)
    y = round((30 - temp_image.height)/2)
    fin_image.paste(temp_image, box = (x, y))
    del temp_image
    return fin_image
```

```python
def resizeim22(image):
    temp_image = image.convert('1')
    #factor = (32 / temp_image.height) if (temp_image.height > temp_image.width) else (32 / temp_image.width)
    factorh = 22 / temp_image.height
    factorw = 22 / temp_image.width
    temp_image = temp_image.resize((round(temp_image.height * factorh), round(temp_image.width * factorw)))
    fin_image = Image.new('1', (30,30), 'white')
    #x = round(15 - (int((x2+1)*factor) - (int((x1-1)*factor)))/2)
    #y = round(15 - (int((y2+1)*factor) - (int((y1-1)*factor)))/2)
    x = round((30 - temp_image.width)/2)
    y = round((30 - temp_image.height)/2)
    fin_image.paste(temp_image, box = (x, y))
    del temp_image
    return fin_image
```

```python
def naive_crop2(image, x1, y1, x2, y2):
    if(((x1>0)and(x2>0))and((y1>0)and(y1>0))):
        temp_image = image.convert('1')
        temp_image = temp_image.crop((x1-1, y1-1, x2+1, y2+1))
        new_image = resizeim(temp_image)
        del temp_image
        return new_image
        #return new_image
    #return new_image
```

```python
def naive_crop_temp(image, x1, y1, x2, y2):
    if(((x1>0)and(x2>0))and((y1>0)and(y1>0))):
        temp_image = image.convert('1')
        temp_image = temp_image.crop((x1-1, y1-1, x2+1, y2+1))
        new_image = resizeim22(temp_image)
        del temp_image
        return new_image
```

```python
def naive_crop(image, x1, y1, x2, y2):
    new_image = Image.new('1',(30, 30), 'white')
    if(((x1>0)and(x2>0))and((y1>0)and(y1>0))):
        x = round(15 - (x2+2 - (x1-2))/2)
        y = round(15 - (y2+2 - (y1-2))/2)
        temp_image = image.convert('1')
        temp_image = temp_image.crop((x1-2, y1-2, x2+2, y2-2))
        new_image.paste(temp_image, box = (x, y))#, (x+x2+2-x1+2), (y+y2+2-y1+2)))
        del temp_image
        return new_image
    elif ((x1>0) and (x2>0)):
        x = round(15 - (x2+2 - (x1-2))/2)
        temp_image = image.convert('1')
        temp_image = temp_image.crop((x1-2, 0, x2+2, 29))
        new_image.paste(temp_image, box = (x, 0))#, (x+x2+2-x1+2), (y+y2+2-y1+2)))
        del temp_image
        return new_image
    elif ((y1>0) and (y2>0)):
        y = round(15 - (y2+2 - (y1-2))/2)
        temp_image = image.convert('1')
        temp_image = temp_image.crop((0, y1-2, 29, y2-2))
        new_image.paste(temp_image, box = (0, y))#, (x+x2+2-x1+2), (y+y2+2-y1+2)))
        del temp_image
        return new_image
        #return new_image
    #return new_image
```

```python
def croptrain_fin(image, imagecounter):
    #image = Image.open(r'C:\Users\NULL\Desktop\BS_project\Image3.tif')
    #image.show()
    for i in range(20):
        for j in range(29):
            for angle in range(-18, 19, 2):
                new_image = image.crop((i*30+3, j*30+3, 27+i*30, 27+j*30))
                fin_image = Image.new('1',(30, 30), 'white')
                new_image = new_image.convert("1")
                x = round((30 - new_image.width)/2)
                y = round((30 - new_image.height)/2)
                fin_image.paste(new_image, box = (x, y))
                
                x1, y1, x2, y2 = naive_cords2(fin_image)
                fin_image = naive_crop_temp(fin_image, x1, y1, x2, y2)
                
                new_image = rotateim(fin_image, angle)
                del fin_image
                
                x1, y1, x2, y2 = naive_cords2(new_image)
                new_image = naive_crop2(new_image, x1, y1, x2, y2)
                if (np.random.randint(0, 10)<2):
                    new_image.save('C:\\Users\\NULL\\Desktop\\BS_project\\check\\' + str(i//2) + '_' + str(imagecounter) + '.tiff', 'TIFF')
                    imagecounter+=1
                else:
                    new_image.save('C:\\Users\\NULL\\Desktop\\BS_project\\train\\' + str(i//2) + '_' + str(imagecounter) + '.tiff', 'TIFF')
                    imagecounter+=1
                del new_image
    del image
    return imagecounter
```
